```python
#5
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("iris.csv")

new_data = data[["Species", "Petal.Length"]]
sns.boxplot(y=new_data["Petal.Length"], x=new_data["Species"], palette="mako")
plt.xlabel("Tipo de flor")
plt.ylabel("Largo del pedal")
plt.title("Boxplot basado en el largo del pétalo por especie")
```

```python
#6
#Basado en el gráfico anterior, se puede determinar que:
#> La especie con el pétalo más largo es virginica.
#> La especie con el pétalo más pequeño es setosa.
#> La especie setosa es la especie que cuenta con más valores atípicos de la muestra.
```
